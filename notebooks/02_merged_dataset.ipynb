# MERGED HYBRID DATASET (STUDENT DATA AND REDDIT DATA)

```python
import pandas as pd

# Load the cleaned Reddit dataset
reddit_clean_path = "../data/processed/cleaned_reddit_dataset.csv"
reddit_clean = pd.read_csv(reddit_clean_path)

# Ensure both datasets have consistent column names
student_df = student_df[['serialized_text', 'Depression']].rename(columns={'serialized_text': 'clean_text', 'Depression': 'binary_label'})

# Combine the Student and Reddit datasets
combined_df = pd.concat([reddit_clean, student_df[['clean_text', 'binary_label']]], ignore_index=True)

# Shuffle the dataset to mix Reddit and Student data
combined_df = combined_df.sample(frac=1).reset_index(drop=True)

# Save the combined dataset to a CSV file
combined_output_path = "../data/processed/combined_hybrid_student_reddit_data.csv"
combined_df.to_csv(combined_output_path, index=False)

# Display the first few rows of the combined dataset
combined_df.head()
```
